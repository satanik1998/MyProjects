# Exploratory Data Analysis

```python
# Group by Aggregate rating and count the number of restaurants
Rating_counts = df['Aggregate rating'].value_counts().sort_index()

# Plot the bar chart
plt.figure(figsize=(10, 5))
Rating_counts.plot(kind='bar')
plt.title('Number of Restaurants by Aggregate Rating')
plt.xlabel('Aggregate Rating')
plt.ylabel('Number of Restaurants')
for i, value in enumerate(Rating_counts):
    plt.text(i, value + 1, str(value), ha='center', va='bottom', fontsize=8)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
    
```

We can see almost a symmetric shape in the curve ,means customers give rating overall .Though the number of 0.0 or No rating is very
high than others ,means a big number of customer didnot give rating. 61 restaurants got 4.9 rating 

```python
# plotting a bar chart for Has Table booking and it's count

Table_Booking= sns.countplot(x = 'Has Table booking',data = df)

for bars in Table_Booking.containers:
    Table_Booking.bar_label(bars)
```

We can see only 1158 restaurants have Table Booking and major student don't have table booking facality.

```python
# plotting a bar chart for Online Delivery Facality and it's count

# Define custom color palette
custom_palette = {'Yes': 'green', 'No': 'red'}

Online_Delivery= sns.countplot(x = 'Has Online delivery',data = df,palette=custom_palette)

for bars in Online_Delivery.containers:
    Online_Delivery.bar_label(bars)
```

We can see only 2451 restaurants have Online Delivery Facality and major student don't have Online Delivery facality.

```python
#5 restaurants with heighst "Average Cost for two people"
Costly_Resaurant = df.sort_values(by='Average Cost for two', ascending=False).head(5)

#Plotting the bar graph
plt.figure(figsize=(15, 8))
plt.bar(Costly_Resaurant['Restaurant Name'], Costly_Resaurant['Average Cost for two'], color='orange', edgecolor='black')
plt.title('Top 5 Restaurants with Highest Average Cost for Two')
plt.xlabel('Restaurant Name')
plt.ylabel('Average Cost for Two')
plt.xticks(rotation=30, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Optional: Add value labels
for i, value in enumerate(Costly_Resaurant['Average Cost for two']):
    plt.text(i, value + 10, f'{value:.0f}', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

From the graph we can see this 5 are the costliest restaurants where in SKye the average cost for two people is 80000

```python
#5 restaurants got heighst votes
Max_Votes = df.sort_values(by='Votes', ascending=False).head(5)

#Plotting the bar graph
plt.figure(figsize=(15, 8))
plt.bar(Costly_Resaurant['Restaurant Name'], Max_Votes['Votes'], color='green', edgecolor='black')
plt.title('Top 5 restaurants got heighst votes')
plt.xlabel('Restaurant Name')
plt.ylabel('Votes')
plt.xticks(rotation=30, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Optional: Add value labels
for i, value in enumerate(Max_Votes['Votes']):
    plt.text(i, value + 10, f'{value:.0f}', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

From the graph we can see this 5 restaurants got heighest votes.Where Skye got the maximum votes.
